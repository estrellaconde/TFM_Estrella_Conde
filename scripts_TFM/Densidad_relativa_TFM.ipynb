```python
#Importamos los paquetes necesarios
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Lectura de los archivos necesarios
df_D0=pd.read_csv("densidad_relativa_arbolado_D0.csv", delimiter=";", decimal=",")
df_DD=pd.read_csv("densidad_relativa_demas_dens.csv", delimiter=";", decimal=",")
```

```python
#Unión de los archivos en un dataframe
df= df_D0.merge(df_DD, how="outer")
df
```

```python
#Optimización del dataframe
df_cod_nom = df.replace({328: "Q. faginea", 
               7150: "Q. ilex", 
               127 :"Pinus sp", 
               2247:"Fraxinus sp ", 
               6206:"C. atlantica"})
df_cod_nom
```

```python
#Limpieza del dataframe
df_nom=df_cod_nom.drop(["id_punto", "densidad_absoluta", "sum_dens_abs"], axis=1)
df_nom
```

```python
df_nom.columns = ["id_visita", "nombre_especie", "densidad_relativa"]
```

```python
#Visualización de las gráficas de densidad relativa por cada punto muestreado
puntos = range (1,19)

fig, axs = plt.subplots(6,3, figsize = (15,23))
plt.subplots_adjust(hspace=0.6)
fig.suptitle("Densidad relativa", fontsize=16, y=0.905)

for i, ax in zip(puntos, axs.ravel()):
    dens_puntos = df_nom["id_visita"] == i
    df_dens_puntos=df_nom[dens_puntos]
    df_dens_puntos.plot(ax=ax, kind = "bar", x="nombre_especie", y="densidad_relativa", legend=False)
    ax.set_title("Punto " + str(i))
    ax.set_xlabel("")
    ax.set_ylabel("%")
    ax.set_ylim(0,105)
```
