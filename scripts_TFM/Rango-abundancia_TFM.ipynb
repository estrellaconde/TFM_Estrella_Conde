```python
#Limpieza de los archivos
df=df_completa2.drop(["id_punto", "cod_especie", "suma_num_tot"], axis=1)
```

```python
#Optimización del dataframe
df_ordenado = df.sort_values("abun_relat", ascending = False)
```

```python
df_ordenado.columns = ["id_visita", "nombre_especie", "abundancia_absoluta", "abundancia_relativa"]
df_ordenado
```

```python
#Visualización de las gráficas de rango-abundancia de los puntos muestreados
puntos = range (1,19)

fig, axs = plt.subplots(6,3, figsize = (20,30))
plt.subplots_adjust(hspace=0.6)
fig.suptitle("Rango-abundancia", fontsize=16, y=0.9)

for i, ax in zip(puntos, axs.ravel()):
    dens_puntos = df_ordenado["id_visita"] == i
    df_dens_puntos=df_ordenado[dens_puntos]
    df_dens_puntos.plot(ax=ax, kind = "bar", x="nombre_especie", y="abundancia_relativa", 
                        legend=False)
    ax.set_title("Punto " + str(i))
    ax.set_xlabel("")
    ax.set_ylabel("%")
    ax.set_ylim(0,41)
```
