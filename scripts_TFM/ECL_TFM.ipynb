```python
#Importamos los paquetes necesarios
import pandas as pd
import numpy as np
```

```python
#Lectura de los archivos que se van a usar
df_ABi=pd.read_csv("df_ABi.csv", delimiter=";", decimal=",")
df_clases=pd.read_csv("clases_DBH_altura.csv", delimiter=";", decimal=",")
df_rege=pd.read_csv("regeneracion_x_visita.csv", delimiter=";", decimal=",")
df_riqueza=pd.read_csv("riqueza_x_visita.csv", delimiter=";", decimal=",")
df_ishannon=pd.read_csv("shannon_indice.csv", delimiter= ";", decimal=",")
```

```python
#Limpieza de los archivos
df_AB=df_ABi.drop(["Unnamed: 0"], axis=1)
df_abi=df_ABi.reindex(columns=["id_visita", "sum_ABi"])
df_abi['sum_ABi'] = df_ABi['sum_ABi'].astype(float, errors = 'raise')
```

```python
df_regen=df_rege.drop(["id_punto", "especie"], axis=1)
regeneracion = df_regen["cod_especie"] == 328
df_regeneracion = df_regen[regeneracion].drop(["cod_especie"],axis=1)
df_regeneracion.reset_index(drop=True, inplace=True)
```

```python
df_shannon=df_ishannon.drop(["id_punto"], axis=1)
```

```python
df_rique=df_riqueza.drop(["id_punto"], axis=1)
```

```python
#Asignación de los valores a una puntuación relacionada con el estado de conservación
df_abi.loc[df_abi["sum_ABi"]>=15, "p_ABi"]=2
df_abi.loc[(df_abi["sum_ABi"]>8) & (df_abi["sum_ABi"]<= 15), "p_ABi"]=1
df_abi.loc[df_abi["sum_ABi"]<=8, "p_ABi"]=0
df_abi
```

```python
df_clases.loc[df_clases["clases_DBH"]>=5, "p_clases_DBH"]=2
df_clases.loc[(df_clases["clases_DBH"]>2) & (df_clases["clases_DBH"]< 5), "p_clases_DBH"]=1
df_clases.loc[df_clases["clases_DBH"]<=2, "p_clases_DBH"]=0
df_clases.loc[df_clases["clases_altura"]>=3, "p_clases_altura"]=2
df_clases.loc[(df_clases["clases_altura"]>1) & (df_clases["clases_altura"]< 3), "p_clases_altura"]=1
df_clases.loc[df_clases["clases_altura"]<=1, "p_clases_altura"]=0
df_clases
```

```python
#Asignación de los valores a una puntuación relacionada con el estado de conservación
df_regeneracion.loc[df_regeneracion["N_plantulas"]>=3, "p_plantulas"]=2
df_regeneracion.loc[(df_regeneracion["N_plantulas"]>1) & (df_regeneracion["N_plantulas"]< 3), "p_plantulas"]=1
df_regeneracion.loc[df_regeneracion["N_plantulas"]<=1, "p_plantulas"]=0
df_regeneracion
```

```python
df_rique.loc[df_rique["riqueza"]>=4, "p_rique"]=2
df_rique.loc[df_rique["riqueza"]==3, "p_rique"]=1
df_rique.loc[df_rique["riqueza"]<=2, "p_rique"]=0
df_rique
```

```python
df_shannon.loc[df_shannon["H"]>=1.48, "p_shannon"]=2
df_shannon.loc[(df_shannon["H"]>0.75) & (df_shannon["H"]< 1.48),"p_shannon"]=1
df_shannon.loc[df_shannon["H"]<=0.74, "p_shannon"]=0
df_shannon
```

```python
#Unión de los archivos en un solo dataframe
df= df_rique.merge(df_abi, how="right").merge(df_regeneracion, how="right").merge(df_clases, how="right")
df
```

```python
#Cálculo del estado de conservación local de cada punto muestreado
ECL=[]
visitas = range (1,17)
for i in visitas:
    id_visita = df["id_visita"] == i
    df_id_visita = df[id_visita]
    x= df_id_visita["p_rique"]+3*df_id_visita["p_ABi"]+df_id_visita["p_plantulas"]+3*df_id_visita["p_clases_DBH"]+3*df_id_visita["p_clases_altura"]+3*2
    ECL.append(x.item())
ECL
```

```python
df_ECL=pd.Series(ECL).to_frame()
df_ECL
# df_ECL.to_csv('df_ECL.csv', sep=';')
```
